```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
df['km/l'] = df['mpg'] * 0.425144
df['hp/km'] = df['km/l'] * df['horsepower']
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,50))
sns.barplot(x = 'horsepower', y = 'name', data = df)
```

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.despine(f)
sns.histplot(
    df,
    x="hp/km", hue="origin",
    multiple="stack",
    palette="light:m_r",
    edgecolor=".3",
    linewidth=.5,
    log_scale=True,
)
ax.set_xticks([200,400 , 600, 800, 1000, 1200, 1400, 1600, 1800, 2000])
```

```python
df.describe()
```

```python
sns.set_theme(style="darkgrid")
sns.kdeplot(
    data=df.query("`hp/km` >= 1000"),
    x="model_year",
    y="acceleration",
    hue="cylinders",
    thresh=.1,
)
```
